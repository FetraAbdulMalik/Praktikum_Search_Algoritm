```python
#Python3 program ti print DFS traversal
# from a given graph
from collections import defaultdict

#kelas ini merepsentasikam sebuah graf yang diarahkan
#menggunakan repsentasi daftar kejadian
class Graph:

	#konstruktor
	def __init__(self):

		#kamus default untuk menyimpan graf
		self.graph = defaultdict(list)

	#fungsi untuk menambhakan tepi graf
	def addEdge(self,u,v):
		self.graph[u].append(v)

	#fungsi yang digunakan DFS

	def DFSUtil(self,v,visited):

		#Tandai node saat ini sebagai sudah dikunjungi
    #dan cetak
		visited.add(v)
		print(v, end = ' ')

		#panggil rekursif untuk semua titik ujung
		#yang berdekatan dengan titik ini
		for neighbour in self.graph[v]:
			if neighbour not in visited:
				self.DFSUtil(neighbour,visited)

	#fungsi untuk melakukan penelusuran DFS. ini menggunakan
	#DFSUTIL() rekursif
	def DFS(self,v):

		#Buat himpunan untuk menyimpan node yang dikunjungi
		visited = set()

		#Panggil fungsi rekursif
		#mencetak penelusuran DFS
		self.DFSUtil(v,visited)
#kode pengguna
if __name__ == "__main__":
	g = Graph()
	g.addEdge(0, 1)
	g.addEdge(0, 2)
	g.addEdge(1, 2)
	g.addEdge(2, 0)
	g.addEdge(2, 3)
	g.addEdge(3, 3)

print ("berikut adalah penelusuran DFS  (di mulai dari node 2)")
#panggilan fungsi
g.DFS(2)
```
